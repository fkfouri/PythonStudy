# Naive Bayes - Duas características

Analisando duas características: **Probabilidade condicional**... neste exemplo, o estado de origem.


P(Comprar | Rio de Janeiro) => *Dado que o usuario é do Rio de Janeiro, qual probabilidade de comprar?*

# Probabilidade condicional de duas varáveis

*Dado que o usuario é de SP e possui entrada menor que $5000, qual probabilidade de comprar?*


\begin{align}
{P(Comprar | São Paulo, <= 5000 )} = \frac{P(Comprar | São Paulo) * P(Comprar | <= 5000 )}{P(Comprar)}
\end{align}


\begin{align}
{P(Comprar | São Paulo, <= 5000 )} = \frac{0.275 * 0.18}{P(Comprar)}
\end{align}
