# Naive Bayes - Duas caractersticas

Analisando duas caractersticas: **Probabilidade condicional**... neste exemplo, o estado de origem.


P(Comprar | Rio de Janeiro) => *Dado que o usuario  do Rio de Janeiro, qual probabilidade de comprar?*

# Probabilidade condicional de duas varveis

*Dado que o usuario  de SP e possui entrada menor que $5000, qual probabilidade de comprar?*


\begin{align}
{P(Comprar | So Paulo, <= 5000 )} = \frac{P(Comprar | So Paulo) * P(Comprar | <= 5000 )}{P(Comprar)}
\end{align}


\begin{align}
{P(Comprar | So Paulo, <= 5000 )} = \frac{0.275 * 0.18}{P(Comprar)}
\end{align}
