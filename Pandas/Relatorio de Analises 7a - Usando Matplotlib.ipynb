## Criando Agrupamentos sem filtro - Matplotlib

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial_valor_m2.csv', sep=';')
dados.head(10)
```

```python
# Estatisticas Descritivas
dados['Valor'].mean()
```

```python
# #quero saber a media por Bairro
# bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
# selecao = dados['Bairro'].isin(bairros)
# dados = dados[selecao]
# dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
type(grupo_bairro)
```

```python
# ver os grupos gerados dentro do DataFrame
grupo_bairro.groups
```

```python
# ver mais informacoes dentro do grupos
for bairro, dados in grupo_bairro:
    print(bairro)
```

```python
for bairro, dados in grupo_bairro:
    print(dados)
```

```python
for bairro, dados in grupo_bairro:
    print(type(dados))
```

```python
for bairro, data in grupo_bairro:
    print('Valor Medio do bairro {} -> {}'.format(bairro, data['Valor'].mean()))
```

```python
grupo_bairro['Valor'].mean().round(2)
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatisticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
# count = quantidade
# mean = media
# std = desvio pardrao
# 25% = primeiro quartil
# 50% = segundo quartil
# 75% = terceiro quartil
```

Note que ha alguns dados estranhos, ex. o valor do aluguel em Botafotgo ser de 4.5 Milhoes e desvio padroa altissimo.

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum'])
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min' : 'Minimo', 'max': 'Maximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize= (20,10))
```

```python
grupo_bairro['Valor'].std().plot.bar(color='blue')
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Medio do Aluguel por Bairro', { 'fontsize' : 22})
```
