## Importando a biblioteca statsmodels

https://www.statsmodels.org/stable/index.html

- uma biblioteca mais estatística
- faz testes de hipoteses
- faz estimações
- tem a coisa mais parecida com a <b>Linguagem R</b>
- tem outputs/testes mais formais

```python
import statsmodels.api as sm
```

## Estimando o modelo com statsmodels

```python
# O statsmodels precisa que seja estimado com constante (o Interceptor)
X_train_com_constante = sm.add_constant(X_train)
X_train_com_constante.head()
```

- coluna const tem haver com Algebra Matricial por tras da estimação do modelo de regressão linear (OLS)
- OLS => Minimos de quadrados Ordinarios (https://pt.wikipedia.org/wiki/M%C3%A9todo_dos_m%C3%ADnimos_quadrados)

```python
modelo_statsmodels = sm.OLS(y_train, X_train_com_constante, hasconst = True).fit()
```

# <font color='red' style='font-size: 30px;'>4.2 Avaliando o Modelo Estimado</font>
<hr style='border: 2px solid red;'>

## Avaliando as estatísticas de teste do modelo

```python
print(modelo_statsmodels.summary())
```

- Estatísticas de resumo
- Nesta análise, dois resultados em especial é o que se deseja analisar: Teste F (F-Statistic) e o Teste T
- R² de 0.805 => valor pode ser mantido ou aumentado tendendo a 1
- R² Ajustado => Considera os graus de liberdade, ou seja, a inclusão de mais variáveis diminui o grau de liberdade.
    - Este dado tem como peculariedade diferente de R², que ele começa a cair conforme for diminuindo o grau de liberdade
- Teste F => Testa a significância dos parametros no modelo se eles conjuntamente forem iguais a zero
    - Prob (F-Statistic) é a aceitação do Teste F. Para valores acima de 0.05 rejeita o modelo, pois eles não são estatisticamente significante, não tem uma relação boa, a regressão nao é boa conjuntamente.
    - Prob (F-Statistic) tem que ser menor ou igual a 0.05.
- Teste T => estatistica por variável.
    - note que log_Dist_Farmacia tem o P>|t| (equivalente ao Prob (F-statistic)) maior que 0.05. <i>o que confirma a suspeita de que log_Dist_Farmacia não era boa. Para que manter se ele não é estatisticamente relevante.</i>
    
- O <b>teste t</b> é utilizado para testar a significância de cada parâmetro individualmente.
- O <b>teste F</b> é utilizado para testar simultaneamente a significância de um conjunto de parâmetros.
- <b>Uma forma de identificar quando os testes t e F indicam que está tudo correto, e então podemos rejeitar a hipótese de que os parâmetros não são significativos individual e conjuntamente.</b>

<p>
    Teste de significância conjunta dos parâmetros <br>
    Prob (F-statistic) <= 0.05 (OK) <br>
Teste de significância individual dos parâmetros <br>
    P>|t| <= 0.05 (OK) <br>
                
</p>

<p><b>Importante para remover variaveis de baixa significancia do modelo e diminuir o CUSTO do modelo</b></p>

## Obtendo o coeficiente de determinação (R²) do modelo estimado com os dados de TREINO

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.score

### Coeficiente de Determinação - R²

O coeficiente de determinação (R²) é uma medida resumida que diz quanto a linha de regressão ajusta-se aos dados. É um valor entra 0 e 1.

$$R^2(y, \hat{y}) = 1 - \frac {\sum_{i=0}^{n-1}(y_i-\hat{y}_i)^2}{\sum_{i=0}^{n-1}(y_i-\bar{y}_i)^2}$$

```python
print('R² = {}'.format(modelo.score(X_train, y_train).round(3)))
```

## Gerando previsões para os dados de TESTE (X_test) utilizando o método *predict()* do objeto "modelo"

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.predict

```python
y_previsto = modelo.predict(X_test)
```

## Obtendo o coeficiente de determinação (R²) para as previsões do nosso modelo

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html#sklearn.metrics.r2_score

```python
print('R² = %s' % metrics.r2_score(y_test, y_previsto).round(3))
```

        <font color='red'>- este tipo de modelo é muito utilizado em estudo economicos, pois eles mendem a eslasticidade de Y em relação as variáveis explicativas. </font>

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>Um aspecto interessante do modelo log-linear, que o tornou muito utilizado nos trabalhos aplicados, é que os coeficientes angulares $\beta_2$ e $\beta_3$ medem as elasticidades de Y em relação a $X_2$ e $X_3$, isto é, a variação percentual de Y correspondente a uma dada variação percentual (pequena) em $X_2$ e $X_3$.</p>
